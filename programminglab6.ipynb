```python

# Sample heart disease dataset (replace with your actual dataset)
heartdata = pd.DataFrame({
    'HeartDisease': [0, 1, 0, 1, 0],  # Example binary values for heart disease
    'BMI': [22.5, 27.0, 24.5, 30.1, 19.5], # Example BMI values
    'AlcoholDrinking': ['yes', 'no', 'no', 'yes', 'yes'],
    'Sex': ['male', 'female', 'female', 'male', 'female'],
    'GenHealth': ['very good', 'good', 'fair', 'good', 'very good'],
    'Stroke': ['yes', 'no', 'no', 'no', 'no'],
    'PhysicalActivity': ['yes', 'no', 'yes', 'no', 'yes'],


})

# Step 1: Use pd.get_dummies() to encode each categorical column with prefixes
alcohol_encoded = pd.get_dummies(heartdata['AlcoholDrinking'], prefix='AlcoholDrinking', dtype=int)
stroke_encoded = pd.get_dummies(heartdata['Stroke'], prefix='Stroke', dtype=int)
physical_activity_encoded = pd.get_dummies(heartdata['PhysicalActivity'], prefix='PhysicalActivity', dtype=int)

# Step 2: Concatenate the encoded DataFrames with the original heartdata DataFrame
heartdata = pd.concat([heartdata, alcohol_encoded, stroke_encoded, physical_activity_encoded], axis=1)




```
